### 2. Całkowicie losowo wygeneruj brakujące dane tak, aby w kolumnach age i chol brakowało po około 10% wartości.


```python
import numpy as np
import random 
rand_data = data.copy()

missing = 0.1
columns = ['age', 'chol']

for col in columns:
    i = 0
    while i < len(rand_data):
        p = random.random()
        if(p<=missing+0.0001):
            rand_data.loc[i, col] = np.nan
        i+=1
```

### 6. Dokonaj imputacji metodami uzupełniania braków poprzednią wartością, interpolacji, MICE i k-najbliższych sąsiadów.


#### *Uzupełnianie poprzednią wartością (Forward Fill)*

```python
forward_fill = rand_data.copy().ffill()#.dropna()
forward_fill = forward_fill.fillna(forward_fill.mean) # (jesli np. w pierwszym rekordzie bedzie brak to forward fill zostawi NaN, nie wypelni go)

forward_fill.describe()
```

#### *Uzupełnianie metodami interpolacji*

```python
interpolation_fill = rand_data.fillna(rand_data.interpolate(method = 'linear'))

interpolation_fill.describe()
```

```python
interpolation_fill2 = rand_data.fillna(rand_data.interpolate(method = 'polynomial', order = 2))

interpolation_fill2.describe()
```

#### *Uzupelnianie MICE*

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(max_iter = 15)
mice_fill = pd.DataFrame(imputer.fit_transform(rand_data), columns=rand_data.columns, index = rand_data.index)

mice_fill.describe()
```

#### *Uzupelnianie metodą k-nn*

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
knn_fill = pd.DataFrame(imputer.fit_transform(rand_data), columns=rand_data.columns, index = rand_data.index)

knn_fill.describe()
```

### 7. Dla każdej z metod policz najmniejszy i największy błąd, a także miary MAE i RMSE.

```python
imputation_methods = {
    "Średnia": mean_imputation,
    "Forward Fill": forward_fill,
    "Interpolacja (Liniowa)": interpolation_fill,
    "Interpolacja (Wielom. 2)": interpolation_fill2,
    "MICE": mice_fill,
    "k-NN (k=5)": knn_fill
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

columns_to_impute = ['age', 'chol']
results = {}

for col in columns_to_impute:
    original_missing_values = data[col][rand_data[col].isna()]
    col_results = {}
    
    for method_name, imputed_data in imputation_methods.items():
        imputed_values = imputed_data[col][rand_data[col].isna()]
        errors = np.abs(original_missing_values - imputed_values)
        min_error = errors.min() #min_blad
        max_error = errors.max() #max_blad
        mae = mean_absolute_error(original_missing_values, imputed_values) #MAE
        rmse = np.sqrt(mean_squared_error(original_missing_values, imputed_values)) #RMSE
        col_results[method_name] = {
            "Najmniejszy Błąd": min_error,
            "Największy Błąd": max_error,
            "MAE": mae,
            "RMSE": rmse
        }
    
    results[col] = pd.DataFrame.from_dict(col_results, orient='index')
```

```python
print(results['age'])
```
